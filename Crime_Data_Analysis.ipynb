```python
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\hites\Downloads\Crime_Data_from_2020_to_Present.csv")
```

```python
df
```

```python
df['Vict Age']= df['Vict Age'].apply(lambda x: x if x>0 else None)
```

```python
df['Weapon Desc']= df['Weapon Desc'].fillna("Unknown")
```

```python
df['Vict Sex']= df['Vict Sex'].fillna("U")
```

```python
df
```

```python
df['Date Reported']= pd.to_datetime(df['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p', errors= 'coerce')
```

```python
df['Date Occurance']= pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p', errors= 'coerce')
```

```python
df
```

```python
df= df.drop_duplicates(subset= 'DR_NO')
```

```python
df['Crime Year']= df['Date Occurance'].dt.year
```

```python
df['Crime Month']= df['Date Occurance'].dt.month
```

```python
df['Crime Day']= df['Date Occurance'].dt.day_name()
```

```python
df
```

```python
df['Crime Time Slot']= df['TIME OCC'].apply(lambda x: "Late Night" if x>=0 and x<600 else "Morning" if x>=600 and x<1200 else "Afternoon" if x>=1200 and x<1800 else "Evening")
```

```python
df['Age Classification']= df['Vict Age'].apply(lambda x: "Minor" if x<18 else "Young Adult" if x>=18 and x<30 else "Adult" if x>=30 and x<50 else "Senior")
```

```python
df
```

```python
area_crime= df.groupby('AREA NAME').size().reset_index(name='Total Crimes')
```

```python
yearly_crime= df.groupby('Crime Year').size().reset_index(name='Crime Count')
```

```python
weapon_usage= df.groupby('Weapon Desc').size().reset_index(name='Weapon Summary')
```

```python
victim_demo= df.groupby(['Vict Age', 'Vict Sex']).size().reset_index(name='Victim Demograph')
```

```python
print(area_crime, yearly_crime, weapon_usage, victim_demo)
```
